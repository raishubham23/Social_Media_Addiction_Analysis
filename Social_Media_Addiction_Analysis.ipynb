# **Univariate Analysis (Distribution of individual features)**

```python
#a. Numeric Columns

numeric_cols = ['Age', 'Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night', 'Mental_Health_Score', 'Conflicts_Over_Social_Media', 'Addicted_Score']
```

```python
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    ax = sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')

    # Add data labels on bars
    for patch in ax.patches:
        height = patch.get_height()
        if height > 0:
            ax.text(patch.get_x() + patch.get_width() / 2, height,
                    int(height), ha='center', va='bottom', fontsize=8)

    plt.tight_layout()
    plt.show()
```

# **Histogram for numerical columns:**

```python
categorical_cols = ['Gender', 'Academic_Level', 'Country', 'Most_Used_Platform', 'Affects_Academic_Performance', 'Relationship_Status']

for col in categorical_cols:
    plt.figure(figsize=(8,5))
    ax = sns.countplot(data=df, x=col, palette='Set2')
    plt.title(f'Distribution of {col}')

    # Add data labels
    for p in ax.patches:
        ax.annotate(f'{p.get_height()}', (p.get_x() + 0.3, p.get_height() + 3))

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

# **Correlation Heatmap**

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

# **Bivariate Analysis**

```python
#Boxplot of Addicted_Score by Gender
plt.figure(figsize=(6,4))
sns.boxplot(x='Gender', y='Addicted_Score', data=df, palette='pastel')
plt.title('Addicted Score by Gender')
plt.show()
```

```python
# Boxplot of Mental_Health_Score by Most_Used_Platform
plt.figure(figsize=(8,5))
sns.boxplot(x='Most_Used_Platform', y='Mental_Health_Score', data=df)
plt.xticks(rotation=45)
plt.title("Mental Health Score by Platform")
plt.show()
```

```python
#Scatter plot: Daily Usage vs Mental Health
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Avg_Daily_Usage_Hours', y='Mental_Health_Score', hue='Gender', palette='Set1')
plt.title("Daily Usage vs Mental Health")
plt.show()
```

# **Pie Chart**

```python
plt.figure(figsize=(6,6))
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#66b3ff','#ff9999'], startangle=90)
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

# **Pair Plot (to see all numerical relationships)**

```python
# Sample small data (for readability of labels)
sample_df = df[numerical_cols].sample(50, random_state=1)  # Avoid using full data for readability

# Create the pairplot
g = sns.pairplot(sample_df)

# Loop over the axes and annotate
for i, j in zip(*np.triu_indices_from(g.axes, 1)):
    ax = g.axes[i, j]
    x_var = numerical_cols[j]
    y_var = numerical_cols[i]
    for k in range(len(sample_df)):
        ax.text(sample_df[x_var].iloc[k], sample_df[y_var].iloc[k], str(k), fontsize=6)

plt.suptitle("Pairplot with Data Labels (Sample)", y=1.02)
plt.show()
```

# **Categorical vs Numeric: Bar plot with means**

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x='Relationship_Status', y='Addicted_Score', data=df, estimator='mean', ci=None, palette='Set2')
plt.title("Average Addicted Score by Relationship Status")

# Add data labels
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.1f}',                # Rounded to 1 decimal
                (p.get_x() + p.get_width() / 2, height),  # Position at center of bar
                ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

# **Interactive Plot with Plotly**

```python
# Import Plotly Express
import plotly.express as px

fig = px.box(df, x="Most_Used_Platform", y="Mental_Health_Score", color="Gender", title="Mental Health Score by Platform")
fig.show()
```

```python
import plotly.express as px

# Create a scatter plot over the box plot
fig = px.box(df,
             x="Most_Used_Platform",
             y="Mental_Health_Score",
             color="Gender",
             title="Mental Health Score by Platform")

# Add individual points with data labels
scatter_trace = px.scatter(
    df,
    x="Most_Used_Platform",
    y="Mental_Health_Score",
    color="Gender",
    text="Mental_Health_Score",  # <== Label to show
    opacity=0.6
).data

fig.add_traces(scatter_trace)

# Update only the scatter plot traces using a selector
fig.update_traces(textposition='top center', selector=dict(mode='markers+text'))
fig.update_layout(showlegend=True)

fig.show()
```
